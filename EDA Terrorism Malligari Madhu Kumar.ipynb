# Cleaning the data

```python
for i in data.columns:
    print(i,end=', ')
```

```python
data.head()
```

```python
#choosing only the data we need as the columns are 135 not feasible to drop
data1=data[["iyear","imonth",'region_txt',"iday","country_txt","city",'propvalue','weaptype1_txt','attacktype1_txt','targtype1_txt','gname','motive','nkill','location','summary','property','longitude','latitude','ndays','nwound','nkillter']]
data1.head()
```

```python
data1.rename(columns={'iyear':'Year','country_txt':'Country','nkill':'killed','nwound':'Wounded','ndays':'Days','gname':'Group Name','imonth':'Month','iday':'Day'},inplace=True)
data1.head()
```

```python
data1.columns
```

```python
#drop duplicate rows
data1=data1.drop_duplicates(keep='first')
```

```python
data1.shape
```

```python
data1['killed'].value_counts()
```

```python
#checking null values
data1.isnull().sum()
```

```python
data1=data1.drop(['propvalue','nkillter','summary'],axis=1)
```

```python

#filling the empty cells
data1["killed"]=data1["killed"].fillna(0)
data1['Wounded']=data1['Wounded'].fillna(0)
data1["casualities"]=data1['killed']+data1['Wounded']
data1['Days']=data1['Days'].fillna(0)
```

```python
data1.isnull().sum()
```

```python

data1_sort=data1.sort_values(by='Country',ascending=True)
data1_sort.head()
```

```python
data1.describe()
```

# Visualising the Data
**Year wise**

```python
attack=data1['Year'].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={'index':'Year','Year':'attacks'}).set_index('Year')
attack.head()
```

```python
attack.plot(kind="bar",figsize=(22,14),fontsize=25)
plt.xlabel("Years",fontsize=25)
plt.ylabel('no.of attacks',fontsize=25)
plt.title('No.of attacks',fontsize=25)
```

```python
wounded=data1[["Year",'Wounded']].groupby('Year').sum()
wounded.head()
```

# #region wise attacks


```python
reg=pd.crosstab(data1.Year,data1.region_txt)
reg.tail()
```

```python
reg.plot(kind='area',figsize=(20,14),fontsize=22,stacked=False)
```

```python
regt=reg.transpose()
regt=regt.sum(axis=1).sort_values()
regt
```

```python
regt.plot(kind='bar',figsize=(20,14),fontsize='21')
plt.xlabel('region',fontsize="20")
plt.ylabel('No.of attacks',fontsize="20")
plt.title('Region wise attacks',fontsize="20")
```
